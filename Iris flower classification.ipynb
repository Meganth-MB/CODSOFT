###  IMPORTING LIBRARIES

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
%matplotlib inline
```

### LOADING THE DATA SET

```python
import pandas as pd
df=pd.read_csv("C:\\Users\\megan\\Downloads\\archive\\IRIS.csv")
print(df.shape)
df.head()
df["species"].value_counts()
```

### VISUALIZING THE DATA SET

```python
import seaborn as sns
df.describe()
sns.pairplot(df,hue='species',height=1.5)
#The following graph explains setosa flowers can be well seperated from non setosa flowers
```

### SEPARATING THE INPUT COLUMN & OUTPUT COLUMN

```python
data=df.values
x=data[:,0:4]
y=data[:,4]
print(x.shape)
print(y.shape)
```

### SPLITING THE DATA TO TRAIN &TEST THE DATA SET

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
print(x_train.shape)
print(x_test.shape)
```

## ML ALGORITHM

### RANDOM FOREST CLASSIFIER

```python
from sklearn.ensemble import RandomForestClassifier
model_RF=RandomForestClassifier()
model_RF.fit(x_train,y_train)
#Training the model
#Make 

y_train_predict=model_RF.predict(x_train)
y_test_predict=model_RF.predict(x_test)


#Training performence
from sklearn.metrics import accuracy_score
model_train_accuracy=accuracy_score(y_train,y_train_predict)


#Test performance
model_test_accuracy=accuracy_score(y_test,y_test_predict)


print('Randam Forest Classifier')
print('Model performance for training set')
print("Accuracy: {:.4f}".format(model_train_accuracy))


print('---------------------------------------------')

print('Model performance for test set')
print("Accuracy: {:.4f}".format(model_test_accuracy))
```
